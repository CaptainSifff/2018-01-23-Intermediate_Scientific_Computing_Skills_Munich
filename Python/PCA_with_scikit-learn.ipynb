# Principal component analysis with scikit-learn


Here we will perform a [principal component analysis](https://en.wikipedia.org/wiki/Principal_component_analysis) on the [Iris flower data set](https://en.wikipedia.org/wiki/Iris_flower_data_set).

Note: scikit learn provides serveral data set including the [iris data set](http://scikit-learn.org/stable/auto_examples/datasets/plot_iris_dataset.html). As as small demostration how to retrieve data we download them from a [web resources](https://archive.ics.uci.edu/ml/datasets/iris). A description of the dataset and its colum names can be found [here](https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.names).

```python
import urllib.request
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
file_name = "iris_data.txt"
urllib.request.urlretrieve(url, file_name)
```

```python
iris_table = pd.read_csv(file_name, names=["sepal length", "sepal_width", "petal_length", "petal_width", "iris_class"])
```

```python
iris_table.head()
```

```python
X_iris = iris_table.loc[: , ["sepal length", "sepal_width", "petal_length", "petal_width"]].values
```

```python
X_iris
```

```python
pca = PCA(n_components=2)
```

```python
principal_components = pca.fit_transform(X_iris)
```

```python
principal_components
```

```python
colors = iris_table.iris_class.apply(
    lambda iris_class: 
    {"Iris-setosa": 0, 
     "Iris-versicolor": 1, 
     "Iris-virginica": 2}[iris_class])
```

```python
plt.scatter(principal_components[:, 0], principal_components[:, 1], c=colors)
```
